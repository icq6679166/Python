```python
import matplotlib.pyplot as plt
x = range(10)
plt.plot(x)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt
```

```python
import numpy as np
ax = np.array(x)
ay = ax**2
# s, o, .,^
plt.plot(ax,ay,'r^--')
```
